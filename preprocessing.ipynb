Можно заметить, что у нас есть $112$ объектов, в которых пропущены какие-то значения. Это всего $3\%$ от общего количества объектов, поэтому я думаю, что можно безопасно от них избавиться

```python
df.dropna(inplace=True)
```

Теперь нормализуем значения в колонках __price__, __rating__, __strength__ и __volume__

```python
for column in ["price", "strength", "volume"]:
    df[column] = (df[column] - df[column].mean()) / df[column].std()

df.head(10)
```

Теперь разберемся с категориальными признаками - __country__, __region__, __grape__, __manufacturer__

Начнем с препроцессинга видов винограда



```python
df["grape"]
```

```python
def process_grapes(grape: str) -> str:
    if grape.endswith("%"):
        return ' '.join(grape.split()[:-1])
    
    return grape


df["grape"] = df["grape"].map(process_grapes)
df.head()
```

```python
df = pd.get_dummies(df, columns=["grape"])
df.head()
```

```python
df = pd.get_dummies(df, columns=["sweetness", "country", "region", "manufacturer"])
df.head()
```

Попробуем дискретизировать значение рейтинга, переведя его в значения $0 \dots 5$

```python
def get_rating_range(rating: float) -> int:
    ranges = [
        (0, 1),
        (1, 2),
        (2, 3),
        (3, 4),
        (4, 5)
    ]
    
    for idx, rang in enumerate(ranges):
        if rang[0] <= rating < rang[1]:
            return idx + 1
    
    return 5 # last range is closed [4, 5] but others are half-open ( [0, 1) for example)
    
```

```python
df["rating"] = df["rating"].map(get_rating_range)
df.tail(10)
```
